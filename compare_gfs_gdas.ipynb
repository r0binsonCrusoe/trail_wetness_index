The purpose of this notebook is to compare the different GFS products and determine which is the best for my use case. I want the most accurate and up to date precipitation and soil moisture data possible for the current day. GDAS is the Global data asssimilation system, which is used to prepare a starting point for the Global forecast system, which actually includes the forecast. Both GDAS and GFS have a surface analysis file. I hope to determine what the difference is between the two. 

They appear to be exactly the same. 

```python
import fsspec
import xarray as xr
```

```python
hour = "00"
date = "20250718"
s3_gfs_nc_path = f"s3://noaa-gfs-bdp-pds/gfs.{date}/{hour}/atmos/gfs.t{hour}z.sfcanl.nc"
fs = fsspec.filesystem("s3", anon=True)

fsspec_caching = {
    "cache_type": "blockcache",  # block cache stores blocks of fixed size and uses eviction using a LRU strategy.
    "block_size": 8
    * 1024
    * 1024,  # size in bytes per block, adjust depends on the file size but the recommended size is in the MB
}

ds_gfs_00 = xr.open_dataset(fs.open(s3_gfs_nc_path, **fsspec_caching), engine="h5netcdf")
ds_gfs_00
```
