```python
from pyrosm import OSM
```

```python
pbf_path = "data/south-carolina-latest.osm.pbf"

trail_filter = {
    "mtb:scale": True,
    "sac_scale": True, 
    "route": ["mtb", "hiking"],
}

osm = OSM(pbf_path) #use get_data function to pull directly from geofabrik

trails = osm.get_data_by_custom_criteria(
    custom_filter=trail_filter,
    filter_type="keep",
)
```

```python
m = trails.explore()

m
```
